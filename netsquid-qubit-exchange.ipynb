## More on NetSquid
it's a discrete-event simulator designed for quantum networks

it allow to:
- Create quantum nodes  
- Connect nodes with quantum channels  
- Model physical effects such as delay and noise  
- Run event-driven simulations  
- Measure fidelity and transmission performance
